```python
import numpy as np
import matplotlib.pyplot as plt
from NumOpt.OptimizationProblem import OptimizationProblem
```

```python
def func(x, beta=1.0):
    assert len(x) == 2, "Input must be a 2-dimensional vector."
    x1, x2 = x[0], x[1]
    f = x1**2 + beta * x2**2
    return f

tol = 1e-6
maxiter = 500
```

```python
for beta in [1.0, 5.0, 15.0]:
    prob = OptimizationProblem(nDim=2, bounds=[(-5,12), (-6, 6)])
    prob.setObjectiveFunction(func, beta)
    history = prob.solve(x0=[10, 4], directionMethod='steepest_descent', stepMethod='backtracking', options={'maxiter':maxiter, 'tol':tol})
    prob.plotSolutionHistory(history)
    # prob.plotFunctionDecreaseHistory(history)
```

```python
# steepest descent + backtracking
for beta in [1.0, 5.0, 15.0]:
    prob = OptimizationProblem(nDim=2, bounds=[(-5,12), (-6, 6)])
    prob.setObjectiveFunction(func, beta)
    history = prob.solve(x0=[10, 4], directionMethod='steepest_descent', stepMethod='strong_wolfe', options={'maxiter':maxiter, 'tol':tol})
    prob.plotSolutionHistory(history)
    # prob.plotFunctionDecreaseHistory(history)
```

```python
# strong_wolfe + conjugate gradient
for beta in [1.0, 5.0, 15.0]:
    prob = OptimizationProblem(nDim=2, bounds=[(-5,12), (-6, 6)])
    prob.setObjectiveFunction(func, beta)
    history = prob.solve(x0=[10, 4], directionMethod='conjugate_gradient', stepMethod='strong_wolfe', options={'maxiter':maxiter, 'tol':tol, 'mu1':1e-4, 'mu2':0.001})
    prob.plotSolutionHistory(history)
    # prob.plotFunctionDecreaseHistory(history)
```

```python
# strong_wolfe + quasi_newton
for beta in [1.0, 5.0, 15.0]:
    prob = OptimizationProblem(nDim=2, bounds=[(-5,12), (-6, 6)])
    prob.setObjectiveFunction(func, beta)
    history = prob.solve(x0=[10, 4], directionMethod='quasi_newton', stepMethod='strong_wolfe', options={'maxiter':maxiter, 'tol':tol, 'mu1':1e-4, 'mu2':0.001})
    prob.plotSolutionHistory(history)
    prob.plotFunctionDecreaseHistory(history)
```
