```python
import numpy as np
import matplotlib.pyplot as plt
from NumOpt.OptimizationProblem import OptimizationProblem
from NumOpt.post_processing import plot_histories
```

```python
# function to test
def func(x):
    assert len(x) == 2, "Input must be a 2-dimensional vector."
    x1, x2 = x[0], x[1]
    f = 0.1*x1**6 - 1.5*x1**4 + 5*x1**2 + 0.1*x2**4 + 3*x2**2 - 9*x2 + 0.5*x1*x2
    return f
```

```python
# PART A
histories = []
methods = ['backtracking', 'strong_wolfe', 'strong_wolfe', 'strong_wolfe', 'strong_wolfe']
mu2 = [1.0, 0.9, 0.5, 0.1, 0.01]

for i,mth in enumerate(methods):
    prob = OptimizationProblem(nDim=2, bounds=[(-3.5,3.5), (0,3)])
    prob.setObjectiveFunction(func)
    hist = prob.solve(x0=[-1.5, 0.1], directionMethod='steepest_descent', stepMethod=mth, options={'maxiter':500, 'tol':1e-6, 'mu2':mu2[i]})
    histories.append(hist)
    prob.plotSolutionHistory(hist)

plt.figure()
for i,h in enumerate(histories):
    plt.plot(h['fval'], '-*', mfc='none', label=methods[i] + " " + str(mu2[i]))
plt.legend()
plt.ylabel('f')
plt.xlabel('iterations')
plt.xlim(-0.1,3.5)
```
