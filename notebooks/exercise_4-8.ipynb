```python
import numpy as np
import matplotlib.pyplot as plt
from NumOpt.OptimizationProblem import OptimizationProblem
from NumOpt.post_processing import plot_histories
```

PART A:
Test the methods for direction search on the Bean function

```python
from NumOpt.functions import beanFunction

# PART A
histories = []
methods = ['steepest_descent', 'conjugate_gradient', 'quasi_newton']
labels = ['Steepest Descent', 'Conjugate Gradient', 'Quasi-Newton']
mu1 = 1e-4
mu2 = 0.01
tol = 1e-6
maxiter = 500

histories = []
for i,mth in enumerate(methods):
    prob = OptimizationProblem(nDim=2, bounds=[(-3.0, 3.0), (-1.0, 3.0)])
    prob.setObjectiveFunction(beanFunction)
    hist = prob.solve(x0=[-1.0, 2.0], directionMethod=mth, stepMethod='strong_wolfe', options={'maxiter':maxiter, 'tol':tol, 'mu1':mu1, 'mu2':mu2})
    histories.append(hist)
    prob.plotSolutionHistory(hist)
```

```python
plot_histories(histories, labels, xlog=(True, True), ylog=(True, True))
```
